```python
import dgl
import torch
```

```python
src = torch.tensor([0, 1, 1, 2, 0])
dst = torch.tensor([0, 0, 1, 1, 2])

g = dgl.graph((src, dst))


features = torch.tensor([[3, 1], [2, 5], [5, -1]]).float()

features.requires_grad_(True)
```

```python
grad_output = torch.ones_like(features)
```

```python
output = dgl.ops.copy_u_min(g, features)
output
```

```python
output.backward(grad_output)
```
